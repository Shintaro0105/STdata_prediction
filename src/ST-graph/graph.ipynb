```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = '/workspaces/STdata_prediction/src/ST-graph/output/main_LA.csv'
data_c = '/workspaces/STdata_prediction/src/ST-graph/output/main_c_bay.csv'
data_s = '/workspaces/STdata_prediction/src/ST-graph/output/main_s_bay.csv'

df = pd.read_csv(data)
df_ = pd.read_csv(data_c)
df__ = pd.read_csv(data_s)

# Plot each metric against Mask_Rate
plt.figure(figsize=(10, 8))

# Plot L1_mean
plt.subplot(2, 2, 1)
plt.plot(df['k'], df['L1_mean'], marker='o', label='main', color='b')
plt.plot(df_['k'], df_['L1_mean'], marker='o', label='main_c', color='r')
plt.plot(df__['k'], df__['L1_mean'], marker='o', label='main_s', color='g')
plt.title('L1_mean')
plt.xlabel('k')
plt.ylabel('L1_mean')
plt.grid(True)
plt.legend()

# Plot L1_std
plt.subplot(2, 2, 2)
plt.plot(df['k'], df['L1_std'], marker='o', label='main', color='b')
plt.plot(df_['k'], df_['L1_std'], marker='o', label='main_c', color='r')
plt.plot(df__['k'], df__['L1_std'], marker='o', label='main_s', color='g')
plt.title('L1_std')
plt.xlabel('k')
plt.ylabel('L1_std')
plt.grid(True)
plt.legend()

# Plot MAE
plt.subplot(2, 2, 3)
plt.plot(df['k'], df['MAE'], marker='o', label='main', color='b')
plt.plot(df_['k'], df_['MAE'], marker='o', label='main_c', color='r')
plt.plot(df__['k'], df__['MAE'], marker='o', label='main_s', color='g')
plt.title('MAE')
plt.xlabel('Mask_Rate')
plt.ylabel('MAE')
plt.grid(True)
plt.legend()

# Plot MAPE
plt.subplot(2, 2, 4)
plt.plot(df['k'], df['MAPE'], marker='o', label='main', color='b')
plt.plot(df_['k'], df_['MAPE'], marker='o', label='main_c', color='r')
plt.plot(df__['k'], df__['MAPE'], marker='o', label='main_s', color='g')
plt.title('MAPE')
plt.xlabel('k')
plt.ylabel('MAPE')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
